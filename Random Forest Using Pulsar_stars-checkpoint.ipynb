```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
pulsar_Data = pd.read_csv('pulsar_stars.csv')
```

```python
Features = [' Mean of the integrated profile',
       ' Standard deviation of the integrated profile',
       ' Excess kurtosis of the integrated profile',
       ' Skewness of the integrated profile', ' Mean of the DM-SNR curve',
       ' Standard deviation of the DM-SNR curve',
       ' Excess kurtosis of the DM-SNR curve', ' Skewness of the DM-SNR curve']
Predicted = ['target_class']
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(pulsar_Data.corr(),annot=True)
```

```python
sns.countplot(x='target_class', data=pulsar_Data)
```

```python
for col in pulsar_Data.columns:
    print(col,':',len(pulsar_Data[col].unique()), 'labels')
```

```python
X = pulsar_Data[Features].values
y = pulsar_Data['target_class'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

```

```python
rf_model = RandomForestClassifier()
```

```python
rf_model.fit(X_train, y_train)
```

```python
rf_prediction = rf_model.predict(X_test)
```

```python
rf_model.score(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
results = cross_val_score(rf_model,X_test,y_test,cv=7)
print('In best case scenario we get a accuracy score of 90% :\n',results)
```

```python
from sklearn import metrics
mse = metrics.mean_squared_error(y_test, rf_prediction)
rmse = np.sqrt(mse)
print(rmse)
```

```python
unique , counts = np.unique(y,return_counts=True)
print(unique,counts)
```

```python
# Evaluations
print('Classification Report: \n')
print(classification_report(y_test,rf_prediction))
print('\nConfusion Matrix: \n')
print(confusion_matrix(y_test,rf_prediction))
```
