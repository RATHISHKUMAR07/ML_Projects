## OpenCV


```python
#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#import OpenCV
```

```python
im = cv2.imread('leaf.jpg')
```

```python
#view image with imshow
plt.imshow(im)
```

###OpenCV reads the channel as **GBR**
###We will convert OpenCV to the RGB channels

```python
img_fix = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

```python
#Scale it to gray and get shape
img_gray = cv2.imread('leaf.jpg', cv2.IMREAD_GRAYSCALE)

#View image
plt.imshow(img_gray, cmap = 'gray')
```

###Resize

```python
#change height width
img_new = cv2.resize(img_fix, (1000,400))
plt.imshow(img_new)
```

```python
#resize with ratio
width_ratio = 0.5
height_ratio =0.5
img_2 = cv2.resize(img_fix, (0,0),
                  img_fix, width_ratio, height_ratio)
plt.imshow(img_2)
```

```python
#shape
img_2.shape
```

###Flipping the Image

```python
#Flip on Horizontal Axis
img_3 = cv2.flip(img_fix,0)
plt.imshow(img_3)
```

```python
#Flip on Vertical Axis
img_4 = cv2.flip(img_fix, 1)
plt.imshow(img_4)
```

```python
#Flip on Horizontal and on Vertical Axis
img_5 = cv2.flip(img_fix, -1)
plt.imshow(img_5)
```

##Draw Shapes on Image 


```python
#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Create Black Image to work 
black_img = np.zeros(shape= (512,512,3), dtype= np.int16)

#Get shape
black_img.shape
```

```python
#Show Image
plt.imshow(black_img)
```

###Circle

```python
#Centre is the first number on x axis and second on y axis
cv2.circle(img = black_img, 
           center = (400,100),
           radius = 50,
           color = (255,0,0),
           thickness = 8)
plt.imshow(black_img)
```

```python
#Filled Circle
#Centre is the first number on x axis and second on y axis
cv2.circle(img = black_img, 
           center = (400,200),
           radius = 50,
           color = (0,255,0),
           thickness = -1)
plt.imshow(black_img)
```

##Rectangle

```python
#The first number is on x- axis & the second on the y- axis
#We need 2 points One for up and one diagonally down
cv2.rectangle(black_img, 
              pt1 = (200,200),
              pt2 = (300,300),
              color = (0, 255, 0),
              thickness = 5)
plt.imshow(black_img)
```

```python
#Filled Rectangle
cv2.rectangle(black_img, 
              pt1 = (200,50),
              pt2 = (300,150),
              color = (137, 79, 213),
              thickness = -1)
plt.imshow(black_img)
```

##Triangle


```python
#Draw Triangle
vertices = np.array([[10, 450],
                     [110, 350],
                     [180, 450]],
                    np.int32)

pts = vertices.reshape(-1,1,2)
cv2.polylines(black_img,
              [pts],
              isClosed = True,
              color = (0, 0, 255),
              thickness = 3)
plt.imshow(black_img)
```

```python
#Filled Triangle
vertices = np.array([[10, 250],
                     [110, 150],
                     [200, 250]],
                    np.int32)

pts = vertices.reshape(-1,1,2)
cv2.fillPoly(black_img,
              [pts],
              color = (255, 167, 201))
plt.imshow(black_img)
```

##Lines

```python
cv2.line(black_img,
         pt1 = (512,0),
         pt2 = (0,512),
         color = (255,0,255),
         thickness = 3)

plt.imshow(black_img)
```

##Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(black_img, 
            text = ' OpenCV',
            org = (210,500),
            fontFace = font,
            fontScale = 2,
            color = (255,255,0),
            thickness = 3,
            lineType = cv2.LINE_AA)

plt.imshow(black_img)
```
