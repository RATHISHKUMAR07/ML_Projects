```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
black_img = np.zeros(shape = (512,512,3),dtype = np.int16)
black_img
```

```python
plt.imshow(black_img)
```

```python
cv2.circle(img = black_img,
          center = (400,200),
          radius = 50,
          color = (255,0,0),
          thickness = 8)
plt.imshow(black_img)
```

```python
cv2.circle(img = black_img,
          center = (400,400),
          radius = 50,
          color = (255,0,0),
          thickness = -2)
plt.imshow(black_img)
```

```python
cv2.rectangle(black_img, 
              pt1 = (200,200),
              pt2 = (300,300),
              color = (0, 255, 0),
              thickness = 5)
plt.imshow(black_img)
```

```python

cv2.rectangle(black_img, 
              pt1 = (200,50),
              pt2 = (300,150),
              color = (137, 79, 213),
              thickness = -1)
plt.imshow(black_img)
```

```python
vertices = np.array([[10, 250],
                     [110, 150],
                     [200, 250]],
                    np.int32)

pts = vertices.reshape(-1,1,2)
cv2.fillPoly(black_img,
              [pts],
              color = (255, 167, 201))
plt.imshow(black_img)
```

```python
cv2.line(black_img,
        pt1 = (512,0),
        pt2 = (0,512),
        thickness = 3,
        color = (255,0,0))
plt.imshow(black_img)
```

```python
cv2.line(black_img,
        pt1 = (128,0),
        pt2 = (0,128),
        thickness = 3,
        color = (255,0,0))
plt.imshow(black_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(black_img, 
            text = ' OpenCV',
            org = (210,500),
            fontFace = font,
            fontScale = 2,
            color = (255,255,0),
            thickness = 3,
            lineType = cv2.LINE_AA)

plt.imshow(black_img)
```
