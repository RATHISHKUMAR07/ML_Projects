##Tune Cascade Classifier
https://docs.opencv.org/2.4/modules/objdetect/doc/cascade_classification.html#cascadeclassifier-detectmultiscale

**detectMultiscale** (input image, **Scale Factor**, **Min Neighbors** )

**Scale Factor** Parameter specifying how much the image size is reduced at each image scale.

**Min Neighbors** Parameter specifying how many neighbors each candidate rectangle should have to retain it.  

##Face Detection

```python
#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Open Image
image = cv2.imread('/content/face_eye.png')

#Fix Image
fix_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
#Create Classifier
face_classifier = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
#Detect Faces
faces = face_classifier.detectMultiScale(fix_img)
```

```python
#No Faces Detected
if faces is():
  print('No Faces Found')
```

```python
def detect_face(fix_img):
  face_rects = faces
  
  for (x,y,w,h) in face_rects:
    cv2.rectangle(fix_img,
                  (x,y),
                  (x+w, y+h),
                  (255,0,0),
                  10)
    return fix_img
```

```python
result = detect_face(fix_img)
plt.imshow(result)
```

##Eyes Detection

```python
#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Open Image
image = cv2.imread('/content/face_eye.png')

#Fix Image
fix_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
#Create Classifier
eye_classifier = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
```

```python
def detect_eyes(fix_img):
  eyes_rects = eye_classifier.detectMultiScale(fix_img)

  for (x,y,w,h) in eyes_rects:
    cv2.rectangle(fix_img,
                  (x,y),
                  (x+w, y+h),
                  (255,0,0,
                   10))
    return detect_eyes
```

```python
result = detect_eyes(fix_img)
plt.imshow(fix_img)
```

##Face and Eyes Detection

```python
#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Open Image
image = cv2.imread('/content/face_eye.png')

#Fix Image
fix_img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
#Create Classifier (Face)
face_classifier = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')

#Create Classifier (Eyes)
eye_classifier = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
```

```python
def detect_faces_eyes(fix_img):
  face_rects = face_classifier.detectMultiScale(fix_img)

  for (x,y,w,h) in face_rects:
    cv2.rectangle(fix_img,
                  (x,y),
                  (x+w, y+h),
                  (255,0,0),
                  7)
    eyes_rects = eye_classifier.detectMultiScale(fix_img)

  for (ix,iy,iw,ih) in eyes_rects:
    cv2.rectangle(fix_img,
                  (ix,iy),
                  (ix+iw, iy+ih),
                  (0,0,255),
                  5)
    return fix_img
    
```

```python
result = detect_faces_eyes(fix_img)
plt.imshow(result)
```
